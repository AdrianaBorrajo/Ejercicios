```python
tab_demo = pd.read_csv('datosige.csv', encoding = 'latin1')
```

```python
tab_demo.head()
```

```python
tab_demo['Tempo'].unique()
```

```python
tab_demo_pivot = pd.pivot_table(tab_demo, index='Tempo',
                                columns=['Indicadores', 'Espazo'],
                                values='DatoN',
                                aggfunc=np.sum)
```

```python
tab_demo_pivot.head()
```

```python
tab_demo_pivot.isnull().sum()
```

```python
tab_demo_pivot.describe()
```

```python
tab_nat = pd.read_csv('datosige2.csv', encoding = 'latin1')
```

```python
tab_nat
```

```python
tab_nat_pivot = pd.pivot_table(tab_nat, index='Tempo',
                               columns=['Movemento natural da poboación','Espazo'],
                               values='DatoN',
                               aggfunc=np.sum)
tab_nat_pivot.head()
```

```python
tab_nat_pivot.isnull().sum()
```

```python
tab_concat = pd.concat([tab_demo_pivot,tab_nat_pivot], axis= 1, sort=False)
tab_concat.head()
```

```python
tab_concat.shape
```

```python
tab_concat['Nacementos'].head()
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(tab_concat.index,tab_concat['Defuncións']['27 Lugo'])
ax.plot(tab_concat.index,tab_concat['Nacementos']['27 Lugo'])
ax.legend(['Defuncions', 'Nacementos'])
```

```python
## autocorrelación
from statsmodels.graphics.tsaplots import plot_acf
fig, ax = plt.subplots(figsize=(8,5))
plot_acf(tab_concat['Defuncións']['27 Lugo'], ax=ax, lags=20)
```

```python
from statsmodels.tsa.ar_model import AutoReg
model = AutoReg(tab_concat['Defuncións']['27 Lugo'], lags=4)
model_fit = model.fit()
```

```python
y = model_fit.predict(len(tab_concat['Defuncións']['27 Lugo']), len(tab_concat['Defuncións']['27 Lugo'])+10)
print(y)
```

```python
plt.figure(figsize=(10,5))
prognose = np.concatenate((tab_concat['Defuncións']['27 Lugo'],y))
plt.plot(range(1,len(prognose)+1), prognose, 'r')
plt.plot(range(1,len(tab_concat['Defuncións']['27 Lugo'])+1), tab_concat['Defuncións']['27 Lugo'], 'b')
xtick_labels = list(range(1981, 2031, 5))
plt.xticks(range(0, 50, 5), xtick_labels)
plt.show()
```
